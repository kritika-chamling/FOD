```python
df = pd.read_csv("Health_heart_experimental.csv")
df1= df.head(10)

sns.lineplot(x="age", y="weightKg", data=df1)

plt.xlabel("Age")
plt.ylabel("WeightKG")
plt.title("Age vs WeightKg")
plt.show()
```

From the given line graph, we can clearly observe that people aged 20–30 years have weights ranging from 45–75 kg. Those aged 30–40 years have weights between 70–90 kg, while the 40–50 years age group shows a weight range of 55–90 kg. Similarly, people aged 50–60 years and 60–70 years have weights ranging from 55–70 kg.

# Bar Plot

```python


df = pd.read_csv("Health_heart_experimental.csv")
df = df.head(5)
#melt() to combine SysBP and DiaBP under BP_Value
df1 = df.melt(id_vars='PatientID', value_vars=['SysBP', 'DiaBP'],
                  var_name='BP_Type', value_name='BP_Value')

# Plot bivariate bar plot
sns.barplot(x='PatientID', y='BP_Value', hue='BP_Type', data=df1)
plt.ylabel('Blood Pressure (mmHg)')
plt.title('Systolic and Diastolic BP')
plt.show()
```

The bar chart shows systolic and diastolic blood pressure readings of first five patients. Systolic pressure is higher than diastolic pressure for all patients, which is normal. Some patients have higher readings than others, showing variation in blood pressure levels.

# Pie Chart

```python
df = pd.read_csv("Health_heart_experimental.csv")

df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

The pie chart shows males account for 50.9% and females for 49.91% of the dataset. Therefore, the number of males and females is nearly equal, with only a 0.09% higher proportion of males.

# Box Plot

```python
df = pd.read_csv("Health_heart_experimental.csv")
sns.boxplot(df['HR'])
```

The box plot shows the distribution of heart rate values. The median heart rate is around 140 bpm, with most values lying between 110 and 170 bpm. The minimum heart rate is about 80 bpm and the maximum is close to 200 bpm, indicating a wide variation in heart rate among individuals.

# Histogram


```python
df = pd.read_csv("Health_heart_experimental.csv")
df = df.sample(20)
plt.hist(df['BMI'])
```

This histogram shows the distribution of BMI values among people. Most BMI values are between about 30 and 40, meaning many people fall in the higher BMI range. Fewer people have BMI values below 25. The height of the bars shows how many people fall into each BMI range.

# Scatter Plot

```python
df = df.tail(20)
sns.scatterplot(x='age', y='heightCm',hue='sex',data=df)
plt.show()
```

From the above scatter plot, we can see that people of the same age can have different heights, indicating high variation in height at every age. The plot also shows that most males are above 150 cm regardless of age, while female heights vary more widely.
