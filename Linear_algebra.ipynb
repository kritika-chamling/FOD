# Assignment of Linear Algebra

```python
import numpy as np
import numpy.linalg as LA
```

# Matrix Operation

```python
A = np.array([
        [2, -1],
        [0, 3]
    ])
B = np.array([
        [1, 2],
        [-3, 4]
    ])
print("A = ",A)
print("B = ",B)
```

```python
# Matrix addition
print("A+B = ", A+B)
```

```python
# Matrix multiplication
result=np.dot(A,B)
print("AB = ", result)
```

```python
# Scalar multiplication 
print("3A=",3*A)
print("2B=",2*B)

# Subtraction
print("3A-2B=",3*A-2*B)
```

# Determinant and Inverse

```python
# Determinant of Matrix A
LA.det(A)
```

Since, the determinant of Matrix A is not equal to 0 so the matrix is invertible. Also, the determinant is np.float64(6.0) i.e 6.0 is stored as numpy 64-bit floating point number.

```python
# Inverse of Matrix A
LA.inv(A)
```

# Eigenvalues and Eigenvectors  

```python
C = np.array([
        [4, 1],
        [-2, 1]
    ])
print("C = ",C)
```

```python
# Eigenvalues and Eigenvector
eigenvalues, eigenvectors = np.linalg.eig(C)

# First eigenvalue and eigenvector
print("\nEigenvalue (first):")
print(eigenvalues[0])
print("Corresponding Eigenvector:")
print(eigenvectors[:, 0])

# Second eigenvalue and eigenvector
print("\nEigenvalue (second):")
print(eigenvalues[1])
print("Corresponding Eigenvector:")
print(eigenvectors[:, 1])
```

# Car Price Prediction Problem

You are given a dataset of cars with the following features:  
- **Number of cylinders (x1)**  
- **Engine size in liters (x2)**  
- **Horsepower (x3)**  

**Given Data:**  
- **Car 1:** x1 = 4, x2 = 1.5, x3 = 100, price = 20  
- **Car 2:** x1 = 6, x2 = 2.0, x3 = 150, price = 30  

**Find:**  
- Price of **Car 3** with features: x1 = 8, x2 = 3.0, x3 = 200



```python
# Feature matrix X
X = np.array([
    [4, 1.5, 100],
    [6, 2.0, 150]
])

# Price vector Y
y = np.array([20, 30])

# Compute coefficient vector beta using pseudo-inverse 
beta = np.dot(np.linalg.pinv(X), y)

print("Beta coefficients:\n", beta)
```

In np.linalg.pinv(X), pinv is the pseudo-inverse of a matrix 
[ i.e pinv = (X.T*X)^âˆ’1*X.T ]
But sometimes (X.T*X) is singular or not invertible so in such cases np.linalg.pinv(X) is used to compute the least-squares solution, which finds the coefficients that minimize the sum of squared differences between predicted and actual values.

```python
# Features of Car 3
car3 = np.array([8, 3.0, 200])

# Price of Car 3
price_car3 = car3 @ beta

print("\nPrice of Car 3:")
print(price_car3)
```
